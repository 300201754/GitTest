```
# firsr importing matplotlib
import matplotlib.pyplot as plt
```

```
distance = [10,20,40,50] # ditance in KM
time=[7,23,34,56] #time in mint
```

```
plt.xlabel("distance")   # for x axis
plt.ylabel("time")        # for y axis
plt.scatter(distance, time, label="this is my points") #  is for points
plt.plot(distance, time, label="only for lines")    #  is for lines
plt.grid(c='green')       # for greads creation   
plt.legend() # to show lables
plt.show()  # to show graph
```

```
# this is cricket data
players = ["dhoni","virat","rohit"]
runs=[100,120,150]
```

```
plt.xlabel("players")
plt.ylabel("run")
plt.bar(players, runs, label="2020 first ckt match")

plt.grid(c='green')
plt.legend()
plt.show()
```

```
plt.xlabel("players")
plt.ylabel("runs")
#plt.bar(players, runs, label="2020 first ckt match")
plt.pie(runs,labels=players, explode=(0,0,0.2), shadow=True, autopct='%1.1f%%')
#plt.grid(c='green')
plt.legend()
plt.show()
```
